```python
def F(x):
    return 2*x[0]**2 + 5*x[1]**2 + x[0]*x[1] - 3*x[0] - x[1]
```

```python
def minhik(h,e):
    x = [1,1]
    x1 = [1,1]
    i=0
    while h>e:
        if F(x)>F([x[0]+h,x[1]]):
            x = [x[0] + h, x[1]]
        elif F(x)>F([x[0]-h, x[1]]):
            x = [x[0]-h , x[1]]
        if F(x)>F([x[0],x[1]+h]):
            x = [x[0],x[1]+h]
        elif F(x)>F([x[0], x[1]-h]):
            x = [x[0], x[1]-h]
        if x==x1:
            h = h/10
        else:
            x1=x


        i = i + 1
        print([i, h, x, F(x)])

    print('')
    print('Полученное оптимальное значение х* = ', x)
    print('Соответствующее значение f(х*) = ', F(x))
```

```python
minhik(0.5,0.0001)
```
